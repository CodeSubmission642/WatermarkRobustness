```python
import sys, os
base_path = os.getcwd()[0:os.getcwd().rfind('Watermark')] + "Watermark/"
sys.path.append(base_path) 

%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np

import tensorflow as tf
from keras import backend as K

from cleverhans.attacks import FastGradientMethod
from cleverhans.utils_keras import KerasModelWrapper

from src.adversarial_main import adversarial_blackbox, zerobit_embed, blackmarks_embed
from src.models import get_deep_cnn_for_cifar, get_lenet_model_for_mnist
from src.preprocess_data import load_cifar_images, load_mnist_images
from src.util import plot_blackbox, merge_histories
from src.callbacks import AdditionalValidationSets
```

```python
sess = tf.Session()
K.set_session(sess)
```

```python
split1 = 15000
split2 = 30000
split3 = 45000
split4 = 60000
```

```python
(x_train, y_train), (x_test, y_test) = load_mnist_images()
```

```python
# Embed the watermark into the model
original_model = get_lenet_model_for_mnist()

original_model, history, trigger = zerobit_embed(model=original_model,
                                     x_train=x_train[:split1],
                                     y_train=y_train[:split1],
                                     x_test=x_test,
                                     y_test=y_test,
                                     sess=sess,
                                     epochs=5,
                                     wm_epochs=5)
```

```python
# Steal the model 
stolen_model = get_lenet_model_for_mnist()

y_pred = original_model.predict(x_train[split1:split2])
stolen_history = AdditionalValidationSets([(trigger['keys'][0], trigger['keys'][1], 'watermark')])
stolen_model.fit(x_train[split1:split2],
                    y_pred,
                    batch_size=64,
                    epochs=5,
                    validation_data=(x_test, y_test),
                    callbacks=[stolen_history],
                    verbose=1)
```

```python
# Train a random model
rnd_model = get_lenet_model_for_mnist()
rnd_history = AdditionalValidationSets([(trigger['keys'][0], trigger['keys'][1], 'watermark')])
rnd_model.fit(x_train[split2:split3],
                    y_train[split2:split3],
                    batch_size=64,
                    epochs=5,
                    validation_data=(x_test, y_test),
                    callbacks=[rnd_history],
                    verbose=1)
```

```python
# Evaluate watermark retention in all models
original_loss = original_model.evaluate(trigger['keys'][0], trigger['keys'][1])
stolen_loss = stolen_model.evaluate(trigger['keys'][0], trigger['keys'][1])
random_loss = rnd_model.evaluate(trigger['keys'][0], trigger['keys'][1])
print("Original: {}, Stolen: {}, Random: {}".format(original_loss[1], stolen_loss[1], random_loss[1]))
```

```python
from keras import losses

# Problem: One number is always turned into one other number in the prediction! 
# Need to add diversity

train_batch_size = 256

def gan_loss_func(y_true, y_pred, y_true_img, y_des_img):
    pred_user = original_model(y_pred)
    pred_stolen = stolen_model(y_pred)
    pred_random = rnd_model(y_pred)

    uniform_classes = np.array(train_batch_size*([0.10]*10)).reshape(-1, 10)
    
    loss_user = losses.mse(pred_user, uniform_classes)
    loss_stolen = losses.mse(pred_stolen, pred_user)
    loss_random = losses.mse(pred_random, pred_user)
    
    loss_img = K.mean(K.abs(y_true - y_pred))

    # Diversity
    pred_var = K.mean(K.softmax(99999999999999*pred_user) - K.softmax(y_des_img))
    
    # Idea: Move to areas with high uncertainty, make stolen loss as low as possible and
    # random loss as high as possible
    
    # 3*loss_img
    total_loss = 2*loss_user + 5*loss_stolen - 2*loss_random + 5*loss_img #+ 15*pred_var

    return total_loss
```

```python
# Try generating better adversarial samples with the GAN (alternatively use gradient descent)
from src.generator_mnist import GAN_MNIST

gan = GAN_MNIST(original_model, 
         batch_size=train_batch_size,
         gan_loss_func=gan_loss_func)
gan.train(train_data=(x_train, y_train),
         epochs=3, 
         sample_stride=50)
```

```python
from keras.utils import to_categorical
def generate_uniform_keys(n_samples):
    return to_categorical(np.random.randint(0, 10, size=n_samples),
                                     num_classes=10).reshape(-1, 10)
```

```python
# Generate keys 
trigger_size = 20000
trigger = gan.generator.predict([x_train[:trigger_size], generate_uniform_keys(trigger_size)])
trigger_pred_y = original_model.predict(trigger)
true_trigger_labels = y_train[:trigger_size]
```

```python
# Select ten keys per class
s_triggers_x, s_triggers_y, s_triggers_true = [], [], []
for i in range(10):
    (trigger_idx,) = np.where(np.argmax(trigger_pred_y, axis=1) == i)
    trigger_idx = trigger_idx[:10]
    if len(trigger[trigger_idx]) > 0:
        s_triggers_x += list(trigger[trigger_idx])
        s_triggers_y += list(trigger_pred_y[trigger_idx])
        s_triggers_true += list(y_train[trigger_idx])
s_triggers_x, s_triggers_y, s_triggers_true = np.array(s_triggers_x), np.array(s_triggers_y), np.array(s_triggers_true)
print("Length: {}".format(len(s_triggers_x)))
```

```python
# Plot adversarial samples
rnd_idx = np.random.randint(len(s_triggers_x), size=10)
to_plot = s_triggers_x[rnd_idx]
labels = np.argmax(s_triggers_true[rnd_idx], axis=1)
new_labels = np.argmax(trigger_pred_y[rnd_idx], axis=1)

fig = plt.figure(figsize=(8, 3))
for i in range(0, 10):
    ax = plt.subplot(2, 5, 1 + i, xticks=[], yticks=[])
    plt.title('{} -> {}'.format(labels[i], new_labels[i]), fontsize=10)
    plt.imshow(np.reshape(to_plot[i], (28, 28)), cmap='gray')

fig.suptitle("Adversarial samples, (top) false (bottom) true", fontsize=16, y=0.06)
plt.show()
```

```python
stolen_loss = stolen_model.evaluate(trigger, trigger_pred_y)
random_loss = rnd_model.evaluate(trigger, trigger_pred_y)
original_loss = original_model.evaluate(trigger, trigger_pred_y)
print("Original Acc: {}, Stolen Acc: {}, Random Acc: {}".format(original_loss[1], stolen_loss[1], random_loss[1]))
```
